```python

import cv2
import numpy as np
import pygame.mixer as mixer
# Initialize the mixer for sound alerts
mixer.init()
sound = mixer.Sound(r"C:\Users\DELL\Desktop\PROJECTS\Drowsiness Using Transfer Learning-CSP\alarm.wav")

# Load the Haar cascades for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Initialize the video capture from the default camera (change index if needed)
cap = cv2.VideoCapture(0)

# Initialize variables
Score = 0
closed_eyes_frames = 0
alarm_playing = False

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

    cv2.rectangle(frame, (0, height - 50), (200, height), (0, 0, 0), thickness=cv2.FILLED)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    for (ex, ey, ew, eh) in eyes:
        eye = frame[ey:ey + eh, ex:ex + ew]
        eye = cv2.resize(eye, (80, 80))
        eye = eye / 255.0  # Normalize the eye image
        eye = np.expand_dims(eye, axis=0)

        # Model prediction
        prediction = model.predict(eye)

        # Determine if eyes are closed
        if prediction[0][0] > 0.30:
            cv2.putText(frame, 'Closed', (10, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255), 1,
                        cv2.LINE_AA)
            Score += 1

            # Play alarm if eyes have been closed for too long
            if Score > 15 and not alarm_playing:
                try:
                    sound.play()
                    alarm_playing = True
                except:
                    pass
        else:
            cv2.putText(frame, 'Open', (10, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255), 1,
                        cv2.LINE_AA)
            Score -= 1

            # Stop alarm when eyes are open
            if alarm_playing:
                try:
                    sound.stop()
                    alarm_playing = False
                except:
                    pass

        # Ensure the score doesn't go below zero
        Score = max(Score, 0)

    cv2.putText(frame, 'Score: ' + str(Score), (100, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255),
                1, cv2.LINE_AA)

    cv2.imshow('Drowsiness Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import pygame.mixer as mixer
from tensorflow.keras.models import load_model

# Load the pre-trained eye state detection model
model = load_model(r"C:\Users\DELL\Desktop\PROJECTS\Drowsiness Using Transfer Learning-CSP\Models\Model.h5")

# Initialize the mixer for sound alerts
mixer.init()
sound = mixer.Sound(r"C:\Users\DELL\Desktop\PROJECTS\Drowsiness Using Transfer Learning-CSP\alarm.wav")

# Load the Haar cascades for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Initialize the video capture from the default camera (change index if needed)
cap = cv2.VideoCapture(0)

# Initialize variables
Score = 0
closed_eyes_frames = 0
alarm_playing = False

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

    cv2.rectangle(frame, (0, height - 50), (200, height), (0, 0, 0), thickness=cv2.FILLED)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    for (ex, ey, ew, eh) in eyes:
        eye = frame[ey:ey + eh, ex:ex + ew]
        eye = cv2.resize(eye, (80, 80))
        eye = eye / 255.0  # Normalize the eye image
        eye = np.expand_dims(eye, axis=0)

        # Model prediction
        prediction = model.predict(eye)

        # Determine if eyes are closed
        if prediction[0][0] > 0.5:  # Adjusted threshold for eye closure
            cv2.putText(frame, 'Closed', (10, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255), 1,
                        cv2.LINE_AA)
            closed_eyes_frames += 1

            # Play alarm continuously if eyes are closed
            if not alarm_playing:
                try:
                    sound.play(-1)  # -1 plays the sound continuously
                    alarm_playing = True
                except:
                    pass
        else:
            cv2.putText(frame, 'Open', (10, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255), 1,
                        cv2.LINE_AA)
            closed_eyes_frames = 0

            # Stop alarm when eyes are open
            if alarm_playing:
                try:
                    sound.stop()
                    alarm_playing = False
                except:
                    pass

    cv2.putText(frame, 'Score: ' + str(Score), (100, height - 20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255),
                1, cv2.LINE_AA)

    cv2.imshow('Drowsiness Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
